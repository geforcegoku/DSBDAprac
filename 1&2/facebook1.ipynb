# Data Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(x="Page total likes", y="Lifetime Engaged Users", data=df)
plt.xlabel("Likes")
plt.ylabel("Engaged Users")
plt.title("Relationship Between Likes and Reach")
plt.show()
```

```python
sns.barplot(x="Category", y="comment", data=df)
plt.xlabel("Post Category")
plt.ylabel("Comments")
plt.title("Average Comments per Post Category")
plt.show()
```
